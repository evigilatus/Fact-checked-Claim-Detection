```python
from glob import glob
import json
from os.path import join
import datefinder
import pandas as pd
from datetime import datetime
```

```python
vclaims_path = '../data/subtask-2b--english/politifact-vclaims'
iclaims_path = '../data/subtask-2b--english/v1/iclaims.queries'
```

```python
def load_vclaims(dir):
    vclaims_fp = glob(f'{dir}/*.json')
    vclaims_fp.sort()
    vclaims = {}
    vclaims_list = []
    for vclaim_fp in vclaims_fp:
        with open(vclaim_fp) as f:
            vclaim = json.load(f)
        vclaims[vclaim['vclaim_id']] = vclaim
        vclaims_list.append(vclaim)
    return vclaims, vclaims_list
```

```python
vclaims, vclaims_list = load_vclaims(vclaims_path)
```

```python
vclaims_date = []

for vclaim in vclaims_list:
    date = list(datefinder.find_dates(vclaim['date'][:30]))[0]
    vclaims_date.append([vclaim['vclaim_id'], date.timestamp()])
    
print(vclaims_date)
    
```

```python
vclaims_date = pd.DataFrame(vclaims_date, columns = ['vclaim_id', 'date_seconds'])
```

```python
vclaims_date.head()
```

```python
vclaims_date.to_csv('vclaims_date.tsv', sep = '\t', index = None)
```

```python
iclaims = pd.read_csv(iclaims_path, sep='\t', names=['iclaim_id', 'iclaim'])
```

```python
iclaims.head()
```

```python
print(len(iclaims))
```

```python
iclaim_ids = list(set(iclaims.iclaim_id.tolist()))
```

```python
print(len(iclaim_ids))
```

```python
iclaims_date = []

for iclaim_id in iclaim_ids:
    iclaim_date = iclaim_id[:8]
    iclaim_date = datetime.strptime(iclaim_date, '%Y%m%d')
    iclaims_date.append([iclaim_id, iclaim_date.timestamp()])
    
print(iclaims_date)
```

```python
iclaims_date = pd.DataFrame(iclaims_date, columns = ['iclaim_id', 'date_seconds'])
```

```python
iclaims_date.head()
```

```python
iclaims_date.to_csv('iclaims_date.tsv', sep = '\t', index = None)
```
